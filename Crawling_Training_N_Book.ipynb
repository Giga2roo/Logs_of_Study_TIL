```python
url = 'https://www.naver.com/nvhaproxy/v1/panels/CULTURE/html'

info = {
    'referer': 'https://www.naver.com/',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36'
}

resp = requests.get(url)
html = BeautifulSoup(resp.text, 'html.parser')
```

```python
intro_list = []
for intro_desc in html.select('strong.title.elss')[:-1]:
    intro_list.append(intro_desc.text)
```

```python
desc_list = []
for descrip in html.select('p.desc')[1:50]:
    desc_list.append(descrip.text)
```

```python
desc_sum = list(zip(intro_list, desc_list))
```
